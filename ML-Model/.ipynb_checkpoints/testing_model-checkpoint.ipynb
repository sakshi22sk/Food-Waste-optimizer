```python
import pandas as pd
import numpy as np
import joblib
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import load_model
from tensorflow.keras.losses import MeanSquaredError
import tensorflow as tf
import pickle
from tensorflow.keras.optimizers import Adam
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np
from sklearn.preprocessing import MinMaxScaler
import warnings
warnings.filterwarnings('ignore')
```

```python

# 📌 Load Datasets
train_path = "food_demand_final.csv"  # Training data
test_path = "food_demand_test.csv"   # Test data

train_data = pd.read_csv(train_path)
test_data = pd.read_csv(test_path)
```

```python
X = test_data.drop(columns=['Food Demand'])  # Features
y = test_data['Food Demand'].values.reshape(-1, 1)  # Target variable reshaped for scaling
```

```python
X.info()
```

```python
# y
```

```python
y.shape
```

```python
scaler_X=joblib.load("scaler_X.pkl")
scaler_y=joblib.load("scaler_y.pkl")
```

```python
X_test_scaled = scaler_X.transform(X)
```

```python
y_test_scaled = scaler_y.transform(y.reshape(-1, 1))
```

```python
time_steps = 10  # Number of past days used for prediction
print(f"X_train_scaled shape: {X_test_scaled.shape}")
print(f"y_train_scaled shape: {y_test_scaled.shape}")

def create_sequences(X, y, time_steps):
    Xs, ys = [], []
    for i in range(len(X) - time_steps):  # Prevent IndexError
        Xs.append(X[i : i + time_steps])
        ys.append(y[i + time_steps - 1])  # Adjust index to stay within bounds
    return np.array(Xs), np.array(ys).reshape(-1, 1)
```

```python
X_test_seq, y_test_seq = create_sequences(X_test_scaled, y_test_scaled, time_steps)
```

```python
X_test_seq.shape
```

```python
y_test_seq.shape
```

```python
# Load model with custom loss function
lstm_model = load_model("lstm_model.h5", custom_objects={'mse': MeanSquaredError()})
# Recompile with the same loss and optimizer
lstm_model.compile(loss=MeanSquaredError(), optimizer=Adam(), metrics=['mae', 'mse'])
```

```python
lstm_predictions = lstm_model.predict(X_test_seq)
```

```python
lstm_predictions = scaler_y.inverse_transform(lstm_predictions)
```

```python
y_test_seq = scaler_y.inverse_transform(y_test_seq) 
```

```python
# Evaluate LSTM model performance
lstm_mae = mean_absolute_error(y_test_seq, lstm_predictions)
lstm_rmse = np.sqrt(mean_squared_error(y_test_seq, lstm_predictions))
lstm_r2 = r2_score(y_test_seq, lstm_predictions)

print(f"LSTM Performance:\nMAE: {lstm_mae:.4f}, RMSE: {lstm_rmse:.4f}, R² Score: {lstm_r2:.4f}")
accuracy = (1 - (lstm_mae/ np.mean(y_test_seq))) * 100 
print(accuracy)
```

```python
# Visualization
plt.scatter(y_test_seq, lstm_predictions, color='blue')
plt.plot(y_test_seq,y_test_seq, color='red', linestyle='dashed')  # Perfect fit line
plt.xlabel("Actual")
plt.ylabel("Predicted")
plt.title("Actual vs Predicted")
plt.show()
```

```python
# Prediction with XGBoost
with open("xgb_model.pkl", "rb") as file:
    xgb_model = pickle.load(file)
```

```python
xgb_predictions = xgb_model.predict(X_test_scaled)
```

```python
xgb_mae = mean_absolute_error(y, xgb_predictions)
xgb_mse = np.mean((y - xgb_predictions) ** 2)
xgb_rmse = np.sqrt(mean_squared_error(y, xgb_predictions))
xgb_r2 = r2_score(y, xgb_predictions)
xgb_accuracy = (1 - (xgb_mae / np.mean(y))) * 100 

print(f"XGBoost Performance:\nMAE: {xgb_mae:.4f}, MSE: {xgb_mse:.4f}, RMSE: {xgb_rmse:.4f}, R² Score: {xgb_r2:.4f}, Accuracy Score: {xgb_accuracy:.4f}")
accuracy = (1 - (xgb_mae / np.mean(y))) * 100 
```

```python
# 1. Scatter Plot
plt.scatter(y, xgb_predictions, color='blue')
plt.plot(y,y, color='red', linestyle='dashed')  # Perfect fit line
plt.xlabel("Actual")
plt.ylabel("Predicted")
plt.title("Actual vs Predicted")
plt.show()
```

```python
# 2. Residual Plot
xgb_predictions_df=pd.DataFrame(xgb_predictions)
residuals = y - xgb_predictions_df
plt.scatter(xgb_predictions_df, residuals, color='blue')
plt.axhline(0, color='red', linestyle='dashed')  # Zero line for reference
plt.xlabel("Predicted")
plt.ylabel("Residuals")
plt.title("Residual Plot")
plt.show()
```

```python
# 3. Histogram of Residuals
# Histogram of residuals
plt.hist(residuals, bins=30, color='blue', edgecolor='black')
plt.xlabel("Residuals (Actual - Predicted)")
plt.ylabel("Frequency")
plt.title("Histogram of Residuals")
plt.show()
```

```python
# 4. Line Plot for Actual vs Predicted (Time Series)
y_flat = y.ravel()  # Flatten the y array to 1D
xgb_predictions_flat = xgb_predictions.ravel()  # Flatten the xgb_predictions array to 1D

# Convert to pandas Series
y_series = pd.Series(y_flat)  
xgb_predictions_series = pd.Series(xgb_predictions_flat)
# Line plot for actual vs predicted over time (for time series data)
plt.plot(y_series.index, y_series, color='blue', label='Actual')  # Actual values
plt.plot(xgb_predictions_series.index, xgb_predictions_series, color='red', label='Predicted')  # Predicted values
plt.xlabel("Time")
plt.ylabel("Food Demand")
plt.title("Actual vs Predicted Food Demand")
plt.legend()
plt.show()
```

```python
# 5. Box Plot of Errors

# Box plot for residuals
plt.boxplot(residuals, vert=False, patch_artist=True, 
            boxprops=dict(facecolor='blue', color='blue'), 
            medianprops=dict(color='red'))  # Optional: to change the color of the median line
plt.xlabel("Residuals (Actual - Predicted)")
plt.title("Box Plot of Residuals")
plt.show()

```

```python
# 6. Performance Metrics Visualization
metrics = ['MAE', 'RMSE', 'R2', 'Accuracy']
values = [xgb_mae, xgb_rmse, xgb_r2, xgb_accuracy]  # Replace with your computed metrics

# Create bar plot
plt.bar(metrics, values, color='blue')

# Annotate each bar with its value
for i, value in enumerate(values):
    plt.text(i, value + 0.01, f'{value:.2f}', ha='center', va='bottom', color='black')

# Add labels and title
plt.ylabel("Metric Value")
plt.title("Performance Metrics")
plt.show()

```

```python
# 7. Predicted vs Actual Distribution Comparison (Density Plot)
plt.figure(figsize=(10, 6))
sns.kdeplot(y, color='blue', label='Actual', shade=True)
sns.kdeplot(xgb_predictions, color='red', label='Predicted', shade=True)
plt.xlabel("Close_forcast")
plt.ylabel("Density")
plt.title("Distribution of Actual vs Predicted Close_forcast")
plt.legend()
plt.show()
```
