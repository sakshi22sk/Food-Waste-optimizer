```python
# Load the transportation cost matrix
cost_df = pd.read_csv("transportation_cost_matrix.csv", index_col=0)

# Create a graph
G = nx.Graph()

# Add nodes (cities + Food Hub)
for city in cost_df.index:
    G.add_node(city)

# Add edges with transportation costs, ignoring -1 values (no direct path)
for i in cost_df.index:
    for j in cost_df.columns:
        if i != j and cost_df.loc[i, j] != -1:
            G.add_edge(i, j, weight=cost_df.loc[i, j])

# Get positions for nodes
pos = nx.spring_layout(G, seed=9)

# Node colors: "Food Hub" is red, others are blue
node_colors = ["red" if city == "Food Hub" else "blue" for city in G.nodes]

# Determine edge colors and thickness based on cost
edge_colors = []
edge_weights = []
min_cost = min(d['weight'] for _, _, d in G.edges(data=True))
max_cost = max(d['weight'] for _, _, d in G.edges(data=True))

for u, v, d in G.edges(data=True):
    cost = d['weight']
    # Normalize cost to a 0-1 range
    norm_cost = (cost - min_cost) / (max_cost - min_cost + 1e-6)
    edge_colors.append((1 - norm_cost, norm_cost, 0))  # Green for low cost, red for high
    edge_weights.append(1 + 4 * norm_cost)  # Thicker edges for higher cost

# Create the figure and axes
fig, ax = plt.subplots(figsize=(10, 8))

# Draw the graph
def draw_graph():
    ax.clear()  # Clear the previous plot

    # Draw nodes
    nx.draw_networkx_nodes(G, pos, node_color=node_colors, node_size=1500, edgecolors="black", ax=ax)

    # Draw edges with colors and variable thickness
    nx.draw_networkx_edges(G, pos, edgelist=G.edges(), edge_color=edge_colors, width=edge_weights, alpha=0.75, ax=ax)

    # Draw labels for cities
    nx.draw_networkx_labels(G, pos, font_size=10, font_weight='bold', font_color='white', ax=ax)

    # Draw edge labels (costs)
    edge_labels = {(u, v): f"{d['weight']}" for u, v, d in G.edges(data=True)}
    nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_size=8, font_color="black", ax=ax)

    ax.set_title("City Transportation Network")
    fig.canvas.draw_idle()  # Redraw the canvas

draw_graph()

plt.show()
```

```python
# Load the transportation cost matrix
cost_df = pd.read_csv("transportation_cost_matrix.csv", index_col=0)

# Replace -1 with NaN (indicating no direct path)
cost_df.replace(-1, float("inf"), inplace=True)

# Create a weighted graph
G = nx.Graph()

# Add nodes (cities + Food Hub)
for city in cost_df.index:
    G.add_node(city)

# Add edges with transportation costs, ignoring 'inf' (no direct path)
for i in cost_df.index:
    for j in cost_df.columns:
        if i != j and cost_df.loc[i, j] != float("inf"):  # Ignore no direct path (-1)
            G.add_edge(i, j, weight=cost_df.loc[i, j])

# Compare shortest path cost vs. direct path cost for all cities
results = []
food_hub = "Food Hub"

for city_number in range(10):  # Cities 0 to 9
    target_city = f"City_{city_number}"

    # Check if there's a direct path
    direct_cost = cost_df.loc[food_hub, target_city]
    if direct_cost == float("inf"):
        direct_cost = None  # No direct path available

    # Compute the shortest path if possible
    try:
        shortest_cost = nx.dijkstra_path_length(G, food_hub, target_city, weight="weight")
    except nx.NetworkXNoPath:
        shortest_cost = None  # No path found

    # Ensure an alternative path exists if no direct path
    if direct_cost is None and shortest_cost is not None:
        direct_cost = "No Direct Path"

    # Check if the shortest path is different from the direct path
    is_different = direct_cost is not None and shortest_cost is not None and shortest_cost != direct_cost

    results.append((target_city, direct_cost, shortest_cost, is_different))

# Convert results to DataFrame
comparison_df = pd.DataFrame(results, columns=["City", "Direct Cost", "Shortest Path Cost", "Different Path Used"])

# Display the comparison
print(comparison_df)

```

```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
import matplotlib.patches as patches

# Load the cost matrix
cost_df = pd.read_csv("transportation_cost_matrix.csv", index_col=0)

# Replace -1 with float("inf") to represent no direct path
cost_df.replace(-1, float("inf"), inplace=True)

# Create a weighted graph
G = nx.Graph()

# Add nodes (cities + Food Hub)
for city in cost_df.index:
    G.add_node(city)

# Add edges with valid transportation costs (ignore -1)
for i in cost_df.index:
    for j in cost_df.columns:
        if i != j and cost_df.loc[i, j] != float("inf"):
            G.add_edge(i, j, weight=cost_df.loc[i, j])

# Function to find shortest path from Food Hub and plot
def find_shortest_path_and_plot(city_number):
    if city_number not in range(10):
        return "Invalid city number! Enter a number between 0 and 9."

    source = "Food Hub"
    target_city = f"City_{city_number}"

    try:
        path = nx.dijkstra_path(G, source, target_city, weight="weight")
        cost = nx.dijkstra_path_length(G, source, target_city, weight="weight")

        # Plot the graph
        pos = nx.spring_layout(G, seed=9)  # Layout for visualization

        # Node colors: red for "Food Hub", yellow for target, blue for others
        node_colors = ["red" if node == source else "yellow" if node == target_city else "blue" for node in G.nodes()]

        # Edge colors: red for path edges, black for others
        edge_colors = ["red" if (u, v) in zip(path, path[1:]) or (v, u) in zip(path, path[1:]) else "black" for u, v in G.edges()]
        edge_widths = [3 if (u, v) in zip(path, path[1:]) or (v, u) in zip(path, path[1:]) else 1 for u, v in G.edges()]

        fig, ax = plt.subplots(figsize=(12, 8))

        nx.draw(G, pos, with_labels=True, node_color=node_colors, edge_color=edge_colors, width=edge_widths, ax=ax, node_size=1500) #increased node size to 1500
        # Draw edge labels (costs) near the nodes
        edge_labels = nx.get_edge_attributes(G, 'weight')
        for edge, label in edge_labels.items():
            u, v = edge
            x = (pos[u][0] + pos[v][0]) / 2
            y = (pos[u][1] + pos[v][1]) / 2

            # Calculate offset to place label near node, not in the middle.
            offset_x = (pos[v][0] - pos[u][0]) * 0.15
            offset_y = (pos[v][1] - pos[u][1]) * 0.15

            ax.text(x + offset_x, y + offset_y, str(label), fontsize=8, color="black", ha='center', va='center', bbox=dict(facecolor='white', alpha=0.7, edgecolor='none'))

        ax.set_title(f"Shortest Path: {source} to {target_city}")

        # Add interactivity: node click highlights neighbors
        def onclick(event):
            if event.inaxes == ax:
                for node, coords in pos.items():
                    if abs(coords[0] - event.xdata) < 0.1 and abs(coords[1] - event.ydata) < 0.1:
                        neighbors = list(G.neighbors(node))
                        highlighted_nodes = [node] + neighbors
                        highlighted_node_colors = ["yellow" if n in highlighted_nodes else color for n, color in zip(G.nodes, node_colors)]
                        nx.draw_networkx_nodes(G, pos, nodelist=G.nodes, node_color=highlighted_node_colors, node_size=1500, ax=ax) #increased node size to 1500
                        fig.canvas.draw_idle()
                        break

        fig.canvas.mpl_connect('button_press_event', onclick)

        plt.show()

        return f"Optimal Path: {'  '.join(path)}\nTotal Cost: {cost}"

    except nx.NetworkXNoPath:
        return f"No path available from {source} to {target_city}!"

# Get user input for city number
city_number = int(input("Enter city number (0-9): "))
print(find_shortest_path_and_plot(city_number))
```
